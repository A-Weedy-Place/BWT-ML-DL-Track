## Introduction:

We had a weather dataset in a csv file.

Our objective was to predict what type of weather it is depending upon the avaliable features.

## Data Cleaning and Preparation:

Our data had 11 columns and 13200 rows
and there were no null values within our data so we didnt had to manage for that.

First i divided the dataset into X and Y.\
Y had only one column "Weather" what we are predicting and rest were in X

then i applied one-hot encoding to all the categorical features (Cloud Cover, Season, Location) since there were not many unique values among them
then i standerdized the numeric features.

I used a preprocessor for this and had a numpy array as an output that i later divided by 80-20 split into training and testing data respectively.

## Model Building

Used 4 models for prediction

1. **Linear Regression**\
It is a simple and highly interpretable model
2. **Decision Tree Classifier**\
It camptures non-linear interactions between features
3. **Random Forest**\
It creates multiple decision trees and aggregates. Usually gives hugh accuracy.
4. **XGBoost**\
This is an advanxed version of gradient boosting \
Ususing gradient boosting to correct the errors of previous models \
Usually used for complez datasets

To all the models the training and testing data as divided before was given and the weather type was predicted.\
But for XGBoost i had to use label encoding for the weather type and it takes in numeic vales instead of strings.
